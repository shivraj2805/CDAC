```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Housing.csv')

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# 1. Histogram of Property Prices
plt.figure(figsize=(12, 6))
sns.histplot(data['price'], bins=30, kde=True, color='blue', alpha=0.5)
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.xscale('log')  # Optional: use log scale for better visualization
plt.show()

# 2. Scatter Plot: Relationship Between Area and Price
plt.figure(figsize=(12, 6))
sns.scatterplot(x='area', y='price', data=data, color='blue', alpha=0.6)
plt.title('Relationship Between Area and Price')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.xscale('log')  # Optional: use log scale for better visualization
plt.yscale('log')  # Optional: use log scale for better visualization
plt.show()

# 3. Bar Plot: Average Price by Number of Bedrooms
avg_price_by_bedrooms = data.groupby('bedrooms')['price'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='bedrooms', y='price', data=avg_price_by_bedrooms, palette='Blues')
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price')
plt.show()
```
